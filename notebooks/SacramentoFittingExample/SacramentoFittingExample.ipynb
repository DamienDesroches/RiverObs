#Sacramento Reach Fitting Example

This notebook shows the end-to-end reach fitting machinery using simulated SWOT data for the Sacramento River.

Below are the Python modules developped for fitting SWOT data over the each:

```python
import numpy as N
from SWOTRiver import SWOTL2
from SWOTRiver import ReachExtractor
from SWOTRiver import RiverObs
from SWOTRiver import FitRiver

# For plotting

%pylab inline
```

# Read the simulation data

In this step, the data are read, a projection to (x,y) coordinates is defined, and the data bounding box is found.

The data locations are given below:

```python
data_dir = '/Volumes/Reservoir/Data/SWOT/'
l2_file = (data_dir +
'swot_intf_line2-9_cycle_0001_pass_0014_w0.23_15dB_dem0.1slp.RightSwath.nc')
db_dir = '../../data/nAmerica_GRWDL_river_topo/'
shape_file_root = db_dir+'nAmerica_GRWDL_river_topo'
```

For this example, only data with classification=1 labels are used (pure water pixels) and the latitude and longitude are assumed to be known from the reference interferogram.

```python
class_list=[1]
lat_kwd='no_layover_latitude'
lon_kwd='no_layover_longitude'

l2 = SWOTL2(l2_file,class_list=class_list,lat_kwd=lat_kwd,lon_kwd=lon_kwd)
```

The heights to be used are the noisy height and the water_height, the true height weighted by the radar response.

```python
h = l2.get('height')
htrue = l2.get('water_height')
```

#Extract the river reaches overlapping the data bounding box

In this case, the reaches are from the North America Global River Width Database for Large Rivers. The data were made by [Tamlin Pavelsky](mailto:pavelsky@unc.edu) and [George Allen](mailto:georgehenryallen@gmail.com) and modified by [Ernesto Rodriguez](mailto:ernesto.rodriguez@jpl.nasa.gov) so that reaches appeared as topologically connected. The results are in the shapefile *nAmerica_GRWDL_river_topo.shp*.

The next step scans through the database and finds all of the overlapping reaches in the data base (only one is found). A small buffer of about 2km is placed about the data bounding box to make sure the river is not cut.

```python
reaches = ReachExtractor(shape_file_root, l2,clip_buffer=0.02)
```

The following step prints out the reach metadata:

```python
reaches[0].metadata
```

#Compute the center line and associate observations with it

In the following step, the centerline is computed along the extracted reach, but the nodes of the centerline are interpolated to be placed every ds=50m. The centerline consists of a set of nodes, each of which has a geolocation and associated data. 

After computing the center line, all points more than max_width/2 away from the centerline are removed. In addition, if a node contains fewer than minobs, it is not considered.

```python
max_width = 400.
ds = 50.
minobs = 10
river_obs = RiverObs(reaches[0],l2.x,l2.y,
                     ds=ds,max_width=max_width,minobs=minobs) 
```

In the following step, the height and true heights are associated with the nodes:

```python
river_obs.add_obs('htrue',htrue)
river_obs.add_obs('h',h)
river_obs.load_nodes(['h','htrue'])
```

Get some rough statistics of the differences in heights for the data associated with each node:

```python
hn_mean = N.array(river_obs.get_node_stat('mean','h'))
hn_median = N.array(river_obs.get_node_stat('median','h'))
hstdn = N.array(river_obs.get_node_stat('stderr','h'))
htn = N.array(river_obs.get_node_stat('mean','htrue'))
sn = N.array(river_obs.get_node_stat('mean','s'))
```

```python
ave = N.mean(hn_mean - htn)*100
err = N.std(hn_mean - htn)*100
print 'Mean statitics:   average: %.1f cm std: %.1f cm'%(ave,err)

ave = N.mean(hn_median - htn)*100
err = N.std(hn_median - htn)*100
print 'Median statitics: average: %.1f cm std: %.1f cm'%(ave,err)
```

Plot the distances from the points to the nodes, and the cross-river normal coordinate:

```python
figsize(10,5)
subplot(1,2,1)
hist(river_obs.d,bins=100,log=False)
xlabel('Distance to node (m)')
xlabel('N obserwations')
grid();
subplot(1,2,2)
hist(river_obs.n,bins=100,log=False)
xlabel('Normal coordinate (m)')
xlabel('N obserwations')
grid();
```

Plot the river geometry, number of observations per node, height vs true height, and a plot of height vs the reach distance downriver.

```python
figsize(10,10)
subplot(2,2,1)
idx = river_obs.populated_nodes
plot(river_obs.centerline.x[idx]/1.e3,river_obs.centerline.y[idx]/1.e3,
     '.',alpha=0.5)
xlabel('x (km)')
ylabel('y (km)')
subplot(2,2,2)
idx = river_obs.populated_nodes
plot(river_obs.centerline.s[idx]/1.e3,river_obs.nobs[idx],'o',alpha=0.5)
grid()
xlabel('Reach (km)')
ylabel('Number of observations')
subplot(2,2,3)
plot(htrue,h,'.',alpha=0.01)
plot([-10,60],[-10,60],'--k',alpha=0.5)
xlim(-10,60)
ylim(-10,60)
grid()
xlabel('h true (m)')
ylabel('h measured (m)')
subplot(2,2,4)
plot(river_obs.s/1.e3,river_obs.htrue,'.',alpha=0.1)
plot(river_obs.s/1.e3,river_obs.h,'.',alpha=0.05)
grid()
xlabel('Reach (km)')
ylabel('Height (m)');
```

Plot the true river height averaged over each node, and the measured height, similarly averaged for mean and median averaging. The gray lines are 2 standard errors away from the truth.

```python
figsize(10,5)
subplot(1,2,1)
plot(sn/1.e3,htn,'.',alpha=0.1)
plot(sn/1.e3,hn_mean,'.',alpha=0.2)
plot(sn/1.e3,htn+2*hstdn,'-k',alpha=0.1)
plot(sn/1.e3,htn-2*hstdn,'-k',alpha=0.1)
xlabel('Reach (km)')
ylabel('Height (m)')
title('Mean')

subplot(1,2,2)
plot(sn/1.e3,htn,'.',alpha=0.1)
plot(sn/1.e3,hn_median,'.',alpha=0.2)
plot(sn/1.e3,htn+2*hstdn,'-k',alpha=0.1)
plot(sn/1.e3,htn-2*hstdn,'-k',alpha=0.1)
xlabel('Reach (km)')
title('Median')
ylabel('Height (m)');
```

#Final step: fit the truth and measurements over the sub-reaches

The fitting algorithm right now implements three different fitting methods, as described in [StatsModels](http://statsmodels.sourceforge.net):

* Ordinary Least Squares (OLS)
* Weighted Least Squares (WLS)
* Robust Linear Models (RLM)

The software will eventually incorporate better outlier rejection at the node level. Right now it has tail trimming, but this is not examined in this example.


In the first step, initialize the fitting object with the observations:

```python
fit = FitRiver(river_obs)
```

## An example

Take the reach between 40km and 50km and fit the measured height and the true height using ordinary unweighted least squares:

```python
smin = 40.e3
smax = 50.e3

h_ols_results = fit.fit_linear(smin,smax,'h',fit='OLS')
yfit_ols = N.dot(fit.X,h_ols_results.params)
h = fit.y

htrue_ols_results = fit.fit_linear(smin,smax,'htrue',fit='OLS')
htrue = fit.y
```

Get the errors in slope and height, convert to suitable units (cm and cm/km) and print:

```python
herror = (h_ols_results.params[1] - htrue_ols_results.params[1])*100
slp_error = (h_ols_results.params[0] - htrue_ols_results.params[0])*1.e5
print('Height error: %.1f cm'%herror)
print(r'Slope error:  %.1f cm/km'%slp_error)
```

Repeat with weighted least squares:

```python

h_wls_results = fit.fit_linear(smin,smax,'h',fit='WLS')
yfit_wls = N.dot(fit.X,h_wls_results.params)

htrue_wls_results = fit.fit_linear(smin,smax,'htrue',fit='WLS')

herror = (h_wls_results.params[1] - htrue_wls_results.params[1])*100
slp_error = (h_wls_results.params[0] - htrue_wls_results.params[0])*1.e5
print('Height error: %.1f cm'%herror)
print(r'Slope error:  %.1f cm/km'%slp_error)
```

Repeat with robust fitting:

```python
h_rlm_results = fit.fit_linear(smin,smax,'h',fit='RLM')
yfit_rlm = N.dot(fit.X,h_rlm_results.params)

htrue_rlm_results = fit.fit_linear(smin,smax,'htrue',fit='RLM')
herror = (h_rlm_results.params[1] - htrue_rlm_results.params[1])*100
slp_error = (h_rlm_results.params[0] - htrue_rlm_results.params[0])*1.e5
print('Height error: %.1f cm'%herror)
print(r'Slope error:  %.1f cm/km'%slp_error)
```

Here is a plot of the three solutions:

```python
figsize(6,6)
plot(fit.s/1.e3,htrue,'-',alpha=0.7,linewidth=2,label='Truth')
plot(fit.s/1.e3,h,'o',alpha=0.2)
plot(fit.s/1.e3,yfit_ols,'-',linewidth=2,label='OLS')
plot(fit.s/1.e3,yfit_wls,'-',linewidth=2,label='WLS')
plot(fit.s/1.e3,yfit_rlm,'-',linewidth=2,label='RLM')
xlabel('Reach (km)')
ylabel('Height (m)')
grid()
legend(loc='best');
```

##Fitting multiple reaches

Now do a more systematic study: start at 30 km reach, and fit successive 10km reaches, displacing each reach by 2.5km. Gather the results and get some statistics. 

```python
herror_ols = []
herror_wls = []
herror_rlm = []
slp_error_ols = []
slp_error_wls = []
slp_error_rlm = []

smin = 30.e3
delta_s = 10.e3
smax = smin + delta_s
step = 2.5e3
sstop = 65.e3

while smax <= sstop:
    
    h_ols_results = fit.fit_linear(smin,smax,'h',fit='OLS')
    htrue_ols_results = fit.fit_linear(smin,smax,'htrue',fit='OLS')
    herror = (h_ols_results.params[1] - htrue_ols_results.params[1])*100
    slp_error = (h_ols_results.params[0] - htrue_ols_results.params[0])*1.e5
    herror_ols.append(herror)
    slp_error_ols.append(slp_error)
    

    h_wls_results = fit.fit_linear(smin,smax,'h',fit='WLS')
    htrue_wls_results = fit.fit_linear(smin,smax,'htrue',fit='WLS')
    herror = (h_wls_results.params[1] - htrue_wls_results.params[1])*100
    slp_error = (h_wls_results.params[0] - htrue_wls_results.params[0])*1.e5
    herror_wls.append(herror)
    slp_error_wls.append(slp_error)
    
    h_rlm_results = fit.fit_linear(smin,smax,'h',fit='RLM')
    htrue_rlm_results = fit.fit_linear(smin,smax,'htrue',fit='RLM')
    herror = (h_rlm_results.params[1] - htrue_rlm_results.params[1])*100
    slp_error = (h_rlm_results.params[0] - htrue_rlm_results.params[0])*1.e5
    herror_rlm.append(herror)
    slp_error_rlm.append(slp_error)
    
    smin += step
    smax += step
```

**Results for Ordinary Least Squares:**

```python
print('Number of sub-reaches: %d\n'%len(herror_ols))
herror_ave = N.mean(herror_ols)
herror_std = N.std(herror_ols)
print('Mean height error: %.1f cm'%herror_ave)
print(' STD height error: %.1f cm'%herror_std)

slp_error_ave = N.mean(slp_error_ols)
slp_error_std = N.std(slp_error_ols)
print(' Mean slope error: %.1f cm/km'%slp_error_ave)
print('  STD slope error: %.1f cm/km'%slp_error_std)
```

**Results for Weighted Least Squares:**

```python
print('Number of sub-reaches: %d\n'%len(herror_ols))
herror_ave = N.mean(herror_wls)
herror_std = N.std(herror_wls)
print('Mean height error: %.1f cm'%herror_ave)
print(' STD height error: %.1f cm'%herror_std)

slp_error_ave = N.mean(slp_error_wls)
slp_error_std = N.std(slp_error_wls)
print(' Mean slope error: %.1f cm/km'%slp_error_ave)
print('  STD slope error: %.1f cm/km'%slp_error_std)
```

**Results for Robust Linear Model:**

```python
print('Number of sub-reaches: %d\n'%len(herror_rlm))
herror_ave = N.mean(herror_rlm)
herror_std = N.std(herror_rlm)
print('Mean height error: %.1f cm'%herror_ave)
print(' STD height error: %.1f cm'%herror_std)

slp_error_ave = N.mean(slp_error_rlm)
slp_error_std = N.std(slp_error_rlm)
print(' Mean slope error: %.1f cm/km'%slp_error_ave)
print('  STD slope error: %.1f cm/km'%slp_error_std)
```
